```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bigCityGermany.csv")
df.info()
```

```python
cities = df[df.Rank <= 3]
```

```python
year_cols = ['1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020', '2021']
melted_df = pd.melt(cities, id_vars=['Rank', 'City', 'State', 'Area in km²', 'first time as a large city'], 
                    value_vars=year_cols, var_name='year', value_name='population')

melted_df 
```

```python
fig = plt.subplots(figsize=(20, 5))
sns.lineplot(melted_df, x = "year", y="population", hue="City", marker = "+").set(title='3 Cities', xlabel='Year', ylabel='Population');
```
