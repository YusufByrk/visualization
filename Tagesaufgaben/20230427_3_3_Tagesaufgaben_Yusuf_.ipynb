```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"]=(10,6)
sns.set_style("whitegrid")
```

```python
diamant = sns.load_dataset("diamonds")
diamant.head()
```

```python
sns.histplot(diamant[diamant['cut']=='Ideal']['price'], kde=True, bins=20, color='blue', edgecolor='black', label='Ideal')
sns.histplot(diamant[diamant['cut']=='Premium']['price'], kde=True, bins=20, color='darkred', edgecolor='black', label='Premium', alpha=0.5)

#ueres Erscheinungsbild
sns.set_style("whitegrid", {"axes.facecolor": ".9"})
sns.despine(left=True, bottom=True)
plt.title("Hufigkeit von Diaanten unterschliedlicher Qualitt abhangig vom Verkaufspreis")
plt.xlabel("Preis")
plt.ylabel("Hufigkeit")

#Legende einfgen
plt.legend(title='Qualitt');
```

***

```python
ax = sns.boxplot(data = diamant, x="cut", y= "carat", palette='PuRd')
ax.set(title="Carat von Diaanten nach Qualittsgrad", ylabel="Carat", xlabel="Qualitt");
```
